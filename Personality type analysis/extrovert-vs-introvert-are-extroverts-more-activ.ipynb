```python
import pandas as pd

df = pd.read_csv('/kaggle/input/extrovert-vs-introvert-behavior-data/personality_dataset.csv')
df['Time_spent_Alone'].isnull().sum()
```

```python
df.fillna({
    'Time_spent_Alone': df['Time_spent_Alone'].mean(),
    'Social_event_attendance': df['Social_event_attendance'].mean(),
    'Going_outside': df['Going_outside'].mean(),
    'Post_frequency': df['Post_frequency'].mean()
}, inplace=True)
```

```python
# Í≤∞Ï∏°Ïπò Ìïú Î≤àÏóê Ï≤òÎ¶¨
df.fillna({
    'Friends_circle_size': df['Friends_circle_size'].median(),
    'Stage_fear': df['Stage_fear'].mode()[0],
    'Drained_after_socializing': df['Drained_after_socializing'].mode()[0]
}, inplace=True)

# Î¨∏ÏûêÏó¥ Ï†ïÎ¶¨ Ìïú Î≤àÏóê Ï≤òÎ¶¨
for col in ['Personality', 'Stage_fear', 'Drained_after_socializing']:
    df[col] = df[col].str.strip().str.lower()
```

```python
df.groupby('Personality')[['Post_frequency', 'Social_event_attendance']].mean()
```

```python
import plotly.express as px

fig = px.box(
    df,
    x='Personality',
    y='Post_frequency',
    title='ÏÑ±Í≤©Î≥Ñ ÏÜåÏÖúÎØ∏ÎîîÏñ¥ Í≤åÏãú ÎπàÎèÑ Î∂ÑÌè¨',
    labels={'Personality': 'ÏÑ±Í≤© Ïú†Ìòï', 'Post_frequency': 'ÏÜåÏÖúÎØ∏ÎîîÏñ¥ Í≤åÏãú ÎπàÎèÑ'}
)

fig.show()
```

```python
df[['Post_frequency', 'Social_event_attendance']].corr()
```

```python
introverts = df[df['Personality'] == 'introvert']
extroverts = df[df['Personality'] == 'extrovert']

print("ÎÇ¥Ìñ•Ìòï ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ:")
print(introverts[['Post_frequency', 'Social_event_attendance']].corr())

print("\nÏô∏Ìñ•Ìòï ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ:")
print(extroverts[['Post_frequency', 'Social_event_attendance']].corr())
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x='Social_event_attendance',
    y='Post_frequency',
    color='Personality',
    title='Ïò§ÌîÑÎùºÏù∏ ÏÇ¨ÌöåÏÑ± vs Ïò®ÎùºÏù∏ ÌëúÌòÑÎ†• (ÏÑ±Í≤©Î≥Ñ)',
    labels={
        'Social_event_attendance': 'ÏÇ¨ÌöåÏ†Å ÌñâÏÇ¨ Ï∞∏ÏÑù ÎπàÎèÑ',
        'Post_frequency': 'ÏÜåÏÖúÎØ∏ÎîîÏñ¥ Í≤åÏãú ÎπàÎèÑ',
        'Personality': 'ÏÑ±Í≤©'
    }
)
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
corr_total = df[['Post_frequency', 'Social_event_attendance']].corr()
corr_intro = introverts[['Post_frequency', 'Social_event_attendance']].corr()
corr_extro = extroverts[['Post_frequency', 'Social_event_attendance']].corr()

# ÏÑúÎ∏åÌîåÎ°Ø ÏÉùÏÑ± (1Ìñâ 3Ïó¥)
fig = make_subplots(rows=1, cols=3, subplot_titles=("Ï†ÑÏ≤¥ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ", "ÎÇ¥Ìñ•Ìòï ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ", "Ïô∏Ìñ•Ìòï ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ"))

# Í∞Å ÌûàÌä∏Îßµ Ï∂îÍ∞Ä
fig.add_trace(
    go.Heatmap(z=corr_total.values,
               x=corr_total.columns,
               y=corr_total.index,
               colorscale='Blues',
               zmin=-1, zmax=1,
               colorbar=dict(title="ÏÉÅÍ¥ÄÍ≥ÑÏàò")),
    row=1, col=1
)

fig.add_trace(
    go.Heatmap(z=corr_intro.values,
               x=corr_intro.columns,
               y=corr_intro.index,
               colorscale='Greens',
               zmin=-1, zmax=1,
               showscale=False),
    row=1, col=2
)

fig.add_trace(
    go.Heatmap(z=corr_extro.values,
               x=corr_extro.columns,
               y=corr_extro.index,
               colorscale='Oranges',
               zmin=-1, zmax=1,
               showscale=False),
    row=1, col=3
)

# Ï†ÑÏ≤¥ Î†àÏù¥ÏïÑÏõÉ ÏÑ§Ï†ï
fig.update_layout(
    title_text="Post Frequency vs Social Event Attendance - Correlation by Personality",
    width=1000,
    height=400
)

fig.show()
```
